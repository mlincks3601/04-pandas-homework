```python
# Dependencies and Setup
import os
import pandas as pd
import numpy as np

# File to Load 
purchase_data_csv= os.path.join("Resources", "purchase_data.csv")

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(purchase_data_csv)
#store into dataframe 
purchase_data_df = pd.DataFrame(purchase_data)
#callback data frame
purchase_data_df
```

## Player Count

* Display the total number of players


```python

#show the total number of players
players_total_amount = purchase_data_df["SN"].count()
#callback to check our total
players_total_amount

print("Totals Players: " + str(players_total_amount))
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#creating a data frame to hold our unique items needed based on avg, counts of values, and the total

unique_df = purchase_data_df.loc[purchase_data_df["Item ID"].unique()].copy()
amount_of_games = unique_df['Item ID'].count()
average_price = purchase_data_df['Price'].mean()
total_games_purchased = purchase_data_df['Purchase ID'].count()
total_revenue = purchase_data_df['Price'].sum()

#now to display our newly gathered info on a summary table and input what we want
summary_data_df = pd.DataFrame({"Number of Products": [amount_of_games],
                           "Average Price": [average_price],
                           "Total Purchases": [total_games_purchased],
                           "Total Revenue": [total_revenue]   
                           })
#now we must place our new data where we want with mapping and formatting
summary_data_df["Average Price"] = summary_data_df["Average Price"].map("${:.2f}".format)
summary_data_df["Total Revenue"] = summary_data_df["Total Revenue"].map("${:.2f}".format)
#callback our table to display it
summary_data_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#using groupby function we will group our data to then find our counts
#find the male players
every_player = purchase_data.loc[purchase_data["Gender"] == "Male"]
male_players = len(every_player)

#find all the female players
every_player = purchase_data.loc[purchase_data["Gender"] == "Female"]
female_players = len(every_player)

#find non-disclosed and others
every_player = purchase_data.loc[purchase_data["Gender"] == "Other / Non-Disclosed"]
nondisclosed_other_players = len(every_player)
#now find the % for each persons len
percentage_of_males = male_players / players_total_amount * 100
percentage_of_females = female_players / players_total_amount * 100
percentage_of_nondisclosed_other_players = nondisclosed_other_players / players_total_amount * 100
#cast our data into a new data frame
Players_Counts_df = pd.DataFrame({"Total Count of Players": [male_players, female_players, nondisclosed_other_players],
                                  "% of Players": [percentage_of_males, percentage_of_females, percentage_of_nondisclosed_other_players],
                                 "Category of Genders": ["Males", "Females", "Other/Non-Disclosed"]}).round(1)
#call our data frame to see what we got
Players_Counts_df
#set our players count to appear in the category of genders index to use later on
players_count_arrange = Players_Counts_df.set_index("Category of Genders")
players_count_arrange.head()
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#convert the age category to a category data type integer
age_int = purchase_data["Age"].astype(str).astype(int)
#now run this code to get my data types
purchase_data.dtypes
#bin our ages 
bin_of_age = [0., 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
age_group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
#categorize using.cut using our age bins
purchase_data["Age_Binning"] = pd.cut(purchase_data["Age"], bin_of_age, labels = age_group_names)
#display our datat frame to see what we got worked
#purchase_data

#Creating a groupby object for my "View Group"
agebin_groups = purchase_data.groupby("Age_Binning")
#now finding how many rows fall into each bin
calc_of_total_players = agebin_groups["Age"].count()
#calculate the total % of players for each age group and divide by total amount of players
percentage_calculation_tp = calc_of_total_players / 576 * 100
age_demographics_df = pd.DataFrame({"% of Players": percentage_calculation_tp,
                                    "Total Count of Age Group": calc_of_total_players}).round(1)
age_demographics_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#group our SN column
SNgroup = purchase_data.groupby("SN")
#count our items in the purchase column
itemcount = SNgroup["Price"].count()
average_purchaseprice_spenders = SNgroup["Price"].mean()
total_purchase_volume = SNgroup["Price"].sum()
top_spenders_df = pd.DataFrame({"Count of Purchases": itemcount,
                                "Average Purchase Price": average_purchaseprice_spenders,
                                "Total Purchase Value": total_purchase_volume}).round(2)
#sort by descending Total purchase value
top_spenders_df = top_spenders_df.sort_values(by ="Total Purchase Value", ascending = False)
#call back and read the first 8 rows
top_spenders_df.head(8)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#group our item id and item name columns
SNgroup2 = purchase_data.groupby(["Item ID", "Item Name"])

#find our unique items then count them 
itemname = SNgroup2["Item Name"].unique()
itemcount2 = SNgroup2["Price"].count()
#find our unique purchases then find the sum 
average_purchase_items = SNgroup2["Price"].count()
total_purchase_items = SNgroup2["Price"].sum()

#create our data frame with our new data
popular_items_df = pd.DataFrame({"Item Purchase Count": itemcount2,
                                 "Item Price": average_purchase_items,
                                 "Total Purchase of Items": total_purchase_items}).round(2)
popular_items_df
```
